# Salary Renegotiation Notebook

This notebook provides a market benchmark for a mid-level `AI Engineer` in Montreal (Fintech) and a contribution dashboard section.

Usage: set `current_salary` in the next code cell, then run the benchmark cell to see suggested targets.

```python
# Parameters — update `current_salary` with your current base salary in CAD (number)
role = "AI Engineer"
industry_filter = "Fintech"
location = "Montreal"
current_salary = None  # e.g. 95000

# Market benchmark — filter by role + (location OR industry) from `market`
bench_market = market[(market["Role"]==role) & ((market["Geography_Level"]==location) | (market.get("Industry")==industry_filter))]

market_avg = bench_market["Avg_Salary"].mean() if not bench_market.empty else None
market_min = bench_market["Min_Salary"].min() if not bench_market.empty else None
market_max = bench_market["Max_Salary"].max() if not bench_market.empty else None

# Experience benchmark — use `experience` if there is a matching Location/Industry/Role row (mid-level rows often have Years_Experience around 3-5)
exp_bench = experience[(experience["Role"]==role) & (experience.get("Industry")==industry_filter) & (experience.get("Location")==location)]
exp_avg = exp_bench["Avg_Salary"].mean() if not exp_bench.empty else None

print("\n=== MARKET BENCHMARK ===")
print(f"Role: {role} — Industry: {industry_filter} — Location: {location}")
print(f"Market Avg: {market_avg:,.0f}" if market_avg is not None else "Market Avg: N/A")
print(f"Market Min: {market_min:,.0f}" if market_min is not None else "Market Min: N/A")
print(f"Market Max: {market_max:,.0f}" if market_max is not None else "Market Max: N/A")
print(f"Experience Avg (filtered): {exp_avg:,.0f}" if exp_avg is not None else "Experience Avg: N/A")

if current_salary is not None and market_avg is not None:
    delta = market_avg - current_salary
    pct = (delta / current_salary) * 100
    print(f"\nYour current salary: {current_salary:,.0f} CAD")
    print(f"Difference vs market avg: {delta:,.0f} CAD ({pct:.1f}%)")
    target = max(market_avg, current_salary * 1.10)  # recommend at least 10% uplift or market avg
    print(f"Suggested target salary: {target:,.0f} CAD")

    # quick plot
    labels = ["You", "Market Avg", "Market Min", "Market Max"]
    values = [current_salary, market_avg, market_min if market_min is not None else 0, market_max if market_max is not None else 0]
    plt.figure(figsize=(6,4))
    sns.barplot(x=labels, y=values, palette=["#107C10","#FFB900","#D13438","#0078D4"])
    plt.ylabel("Salary (CAD)")
    plt.title("Salary vs Market Benchmark")
    plt.tight_layout()
    plt.show()
else:
    print("\nSet `current_salary` (number) and re-run this cell to see personalized suggestions.")
```

```python
# =============================
# VIS 1 — KPI CARDS (Matplotlib)
# =============================
kpi_values = {
    "Global Avg Salary": market[market["Geography_Level"]=="Worldwide"]["Avg_Salary"].mean(),
    "Canada Avg Salary": market[market["Geography_Level"]=="Canada"]["Avg_Salary"].mean(),
    "Montreal Avg Salary": market[market["Location"]=="Montreal"]["Avg_Salary"].mean(),
    "Fintech Montreal Avg": market[(market["Location"]=="Montreal") & (market["Industry"]=="Fintech")]["Avg_Salary"].mean()
}

fig, axes = plt.subplots(1,4, figsize=(20,5))
colors = ["#0078D4", "#0078D4", "#0078D4", "#FFB900"]

for ax, (title, value), color in zip(axes, kpi_values.items(), colors):
    display_val = f"${value:,.0f}" if pd.notna(value) else "N/A"
    ax.text(0.5, 0.5, display_val, fontsize=28, ha='center', va='center', color=color)
    ax.set_title(title, fontsize=16)
    ax.axis("off")
# =============================
# PLOTLY EXPORT — Interactive charts & handout generation
# =============================
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import os
from PIL import Image
os.makedirs('handout', exist_ok=True)
def _safe_write(fig, path):
    try:
        fig.write_image(path, engine='kaleido')
    except Exception as e:
        print(f'Skipping image write for {path}: {e}')

# 1) KPI indicators (normalize currencies to CAD and show counts)
usd_to_cad = 1.35
market_k = market.copy()
if 'Currency' in market_k.columns:
    market_k['Avg_Salary_CAD'] = market_k.apply(lambda r: r['Avg_Salary'] * usd_to_cad if str(r.get('Currency')).upper().strip() == 'USD' else r['Avg_Salary'], axis=1)
else:
    market_k['Avg_Salary_CAD'] = market_k['Avg_Salary']
kpi_items = [
    ('Global Avg Salary', market_k[market_k.get('Geography_Level','')=='Worldwide']),
    ('Canada Avg Salary', market_k[market_k.get('Geography_Level','')=='Canada']),
    ('Montreal Avg Salary', market_k[market_k.get('Location','')=='Montreal']),
    ('Fintech Montreal Avg', market_k[(market_k.get('Location','')=='Montreal') & (market_k.get('Industry','')=='Fintech')]),
]
fig_kpis = make_subplots(rows=1, cols=4, specs=[[{'type':'indicator'},{'type':'indicator'},{'type':'indicator'},{'type':'indicator'}]])
for i,(title,df) in enumerate(kpi_items, start=1):
    val = float(df['Avg_Salary_CAD'].mean()) if (df is not None and not df.empty) else 0
    count = len(df) if (df is not None) else 0
    subtitle = f'{count} rows • CAD'
    fig_kpis.add_trace(go.Indicator(mode='number', value=val, title={'text': f"{title}<br><span style=\"font-size:12px;color:gray\">{subtitle}</span>"}), row=1, col=i)
fig_kpis.update_layout(height=220)
fig_kpis.write_html('handout/kpis.html', include_plotlyjs=True)
_safe_write(fig_kpis, 'handout/kpis.png')

# 2) Geography Avg Salary (interactive)
ai_geo = market[market['Role']=='AI Engineer']
fig_geo = px.bar(ai_geo, x='Avg_Salary', y='Geography_Level', orientation='h', color='Industry', title='Avg AI Engineer Salary by Geography (CAD)')
fig_geo.update_layout(height=400)
fig_geo.write_html('handout/geo.html', include_plotlyjs=True)
_safe_write(fig_geo, 'handout/geo.png')

# 3) Experience progression (interactive)
exp_fin = experience[(experience['Industry']=='Fintech') & (experience['Role']=='AI Engineer')]
if not exp_fin.empty:
    fig_exp = px.line(exp_fin, x='Years_Experience', y='Avg_Salary', color='Location', markers=True, title='Salary Progression by Experience (Fintech, CAD)')
    fig_exp.write_html('handout/exp_progression.html', include_plotlyjs=True)
    _safe_write(fig_exp, 'handout/exp_progression.png')
else:
    fig_exp = None

# 4) Salary vs Experience scatter (interactive)
exp_df = experience[(experience['Role']=='AI Engineer') & (experience['Location']=='Montreal')].copy()
def years_to_num(s):
    try:
        if isinstance(s,(int,float)): return float(s)
        s=str(s).strip();
        if '-' in s: a,b=s.split('-'); return (float(a)+float(b))/2
        if '+' in s: return float(s.replace('+',''))
        return float(s)
    except: return None
exp_df['Years_num']=exp_df['Years_Experience'].apply(years_to_num)
exp_plot = exp_df.dropna(subset=['Years_num','Avg_Salary'])
if not exp_plot.empty:
    fig_scatter = px.scatter(exp_plot, x='Years_num', y='Avg_Salary', color='Industry', size_max=12, title='Salary vs Experience — AI Engineer (Montreal)')
    fig_scatter.update_traces(marker=dict(size=10))
    fig_scatter.write_html('handout/salary_vs_exp.html', include_plotlyjs=True)
    _safe_write(fig_scatter, 'handout/salary_vs_exp.png')
else:
    fig_scatter = None

# 5) Total comp distribution (interactive)
tc = exp_23[(exp_23.get('Role')=='AI Engineer') & (exp_23.get('Location')=='Montreal')]
if not tc.empty:
    fig_tc = px.violin(tc, x='Industry', y='Total_Comp_With_Bonus', box=True, points='all', title='Total Compensation Distribution (2–3 yrs, Montreal)')
    fig_tc.write_html('handout/total_comp.html', include_plotlyjs=True)
    _safe_write(fig_tc, 'handout/total_comp.png')
else:
    fig_tc = None

# 6) Percentiles chart (interactive) — reuse pct_df
import numpy as np
pct_df = exp_23[(exp_23.get('Role')=='AI Engineer') & (exp_23.get('Location')=='Montreal')]
if not pct_df.empty:
    values = pct_df['Avg_Salary'].dropna().astype(float)
    p10,p25,p50,p75,p90 = np.percentile(values, [10,25,50,75,90])
    fig_pct = px.bar(x=['10th','25th','50th','75th','90th'], y=[p10,p25,p50,p75,p90], labels={'x':'Percentile','y':'Salary (CAD)'}, title='Salary Percentiles — AI Engineer (Montreal, 2–3 yrs)')
    fig_pct.write_html('handout/percentiles.html', include_plotlyjs=True)
    _safe_write(fig_pct, 'handout/percentiles.png')
else:
    fig_pct = None

# 7) AI Market Share Donut (Plotly) — replace previous industry pie
# Uses the provided `industry` dataset (AI_Demand_By_Industry_2026.csv)
fig = px.pie(
    industry,
    names="Industry",
    values="Market_Share_Pct",
    hole=0.4,
    color="Industry",
    color_discrete_map={i:"#FFB900" if "Fintech" in i else "#0078D4" for i in industry["Industry"]}
)
fig.update_layout(title_text="AI Market Share by Industry (2026)")
# export interactive and static versions for the handout
fig.write_html('handout/industry_share.html', include_plotlyjs=True)
_safe_write(fig, 'handout/industry_share.png')
# show in-notebook (when running interactively)
try:
    fig.show()
except Exception:
    pass

# 8) Experience vs Position — Fintech vs All (external CSV)
exp_pos_path = 'data/exp_position_sector.csv'
if os.path.exists(exp_pos_path):
    exp_pos = pd.read_csv(exp_pos_path)
    # ensure numeric Years
    exp_pos['Years'] = pd.to_numeric(exp_pos['Years'], errors='coerce')
    fig_exppos = px.line(exp_pos, x='Years', y='Avg_Salary', color='Sector', line_dash='Position', markers=True, title='Salary by Years vs Position — Fintech vs All')
    fig_exppos.write_html('handout/exp_vs_position.html', include_plotlyjs=True)
    _safe_write(fig_exppos, 'handout/exp_vs_position.png')
else:
    print('Missing data/exp_position_sector.csv — chart skipped')

# Build 3 combined HTMLs with embedded content (printable)
def embed_chart_content(html_path):
    """Read and embed the full HTML content for printing"""
    try:
        with open(html_path, 'r', encoding='utf-8') as f:
            content = f.read()
            # Extract just the body content if it exists
            if '<body>' in content:
                start = content.find('<body>') + 6
                end = content.find('</body>')
                if end > start:
                    content = content[start:end]
            return content
    except:
        return f'<p style="color:red">Chart not found: {html_path}</p>'

# Dashboard 1: Benchmark AI market
html_parts_1 = ['<html><head><meta charset="utf-8"><title>Benchmark Dashboard</title><style>body{font-family:Arial,sans-serif;margin:20px}h2{page-break-before:always;margin-top:40px;color:#0078D4}h2:first-of-type{page-break-before:auto}@media print{h2{page-break-before:always}}</style></head><body><h1>Salary Dashboard - Benchmark</h1>']
chart_names_1 = ['kpis','geo','industry_share','exp_progression','salary_vs_exp','total_comp','percentiles']
for name in chart_names_1:
    path = f'outputs/handout/{name}.html'
    if os.path.exists(path):
        html_parts_1.append(f'<h2>{name.replace("_"," ").title()}</h2>')
        html_parts_1.append(embed_chart_content(path))
html_parts_1.append('</body></html>')
combined_html_1 = ''.join(html_parts_1)
with open('outputs/handout/salary_handout_benchmark.html','w',encoding='utf-8') as f: f.write(combined_html_1)

# Dashboard 2: Contribution Dashboard
html_parts_2 = ['<html><head><meta charset="utf-8"><title>Contribution Dashboard</title><style>body{font-family:Arial,sans-serif;margin:20px}h2{page-break-before:always;margin-top:40px;color:#0078D4}h2:first-of-type{page-break-before:auto}@media print{h2{page-break-before:always}}</style></head><body><h1>Salary Dashboard - Contribution</h1>']
chart_names_2 = ['value_propositions_radar','montreal_2_3_years','montreal_companies','salary_position_evolution','technical_skills_evolution']
for name in chart_names_2:
    path = f'outputs/handout/{name}.html'
    if os.path.exists(path):
        html_parts_2.append(f'<h2>{name.replace("_"," ").title()}</h2>')
        html_parts_2.append(embed_chart_content(path))
html_parts_2.append('</body></html>')
combined_html_2 = ''.join(html_parts_2)
with open('outputs/handout/salary_handout_contribution.html','w',encoding='utf-8') as f: f.write(combined_html_2)

# Dashboard 3: Negotiation Summary
html_parts_3 = ['<html><head><meta charset="utf-8"><title>Negotiation Dashboard</title><style>body{font-family:Arial,sans-serif;margin:20px}h2{page-break-before:always;margin-top:40px;color:#0078D4}h2:first-of-type{page-break-before:auto}@media print{h2{page-break-before:always}}</style></head><body><h1>Salary Dashboard - Negotiation</h1>']
chart_names_3 = ['salary_targets_gauge','talking_points','strategic_questions','soft_skills_growth']
for name in chart_names_3:
    path = f'outputs/handout/{name}.html'
    if os.path.exists(path):
        html_parts_3.append(f'<h2>{name.replace("_"," ").title()}</h2>')
        html_parts_3.append(embed_chart_content(path))
html_parts_3.append('</body></html>')
combined_html_3 = ''.join(html_parts_3)
with open('outputs/handout/salary_handout_negotiation.html','w',encoding='utf-8') as f: f.write(combined_html_3)

print('✅ Saved 3 printable dashboards:')
print('  - outputs/handout/salary_handout_benchmark.html')
print('  - outputs/handout/salary_handout_contribution.html')
print('  - outputs/handout/salary_handout_negotiation.html')
# Generate 3 separate PDFs (one for each dashboard)
try:
    from PIL import Image
    
    # PDF 1: Benchmark Dashboard
    pngs_1 = [p for p in ['outputs/handout/kpis.png','outputs/handout/geo.png','outputs/handout/industry_share.png','outputs/handout/exp_progression.png','outputs/handout/salary_vs_exp.png','outputs/handout/total_comp.png','outputs/handout/percentiles.png'] if os.path.exists(p)]
    if pngs_1:
        imgs_1 = [Image.open(p).convert('RGB') for p in pngs_1]
        imgs_1[0].save('outputs/handout/salary_handout_benchmark.pdf', save_all=True, append_images=imgs_1[1:])
        print(f'✅ PDF 1: salary_handout_benchmark.pdf ({len(pngs_1)} pages)')
    
    # PDF 2: Contribution Dashboard
    pngs_2 = [p for p in ['outputs/handout/value_propositions_radar.png','outputs/handout/montreal_2_3_years.png','outputs/handout/montreal_companies.png','outputs/handout/salary_position_evolution.png','outputs/handout/technical_skills_evolution.png'] if os.path.exists(p)]
    if pngs_2:
        imgs_2 = [Image.open(p).convert('RGB') for p in pngs_2]
        imgs_2[0].save('outputs/handout/salary_handout_contribution.pdf', save_all=True, append_images=imgs_2[1:])
        print(f'✅ PDF 2: salary_handout_contribution.pdf ({len(pngs_2)} pages)')
    
    # PDF 3: Negotiation Dashboard
    pngs_3 = [p for p in ['outputs/handout/salary_targets_gauge.png','outputs/handout/talking_points.png','outputs/handout/strategic_questions.png','outputs/handout/soft_skills_growth.png'] if os.path.exists(p)]
    if pngs_3:
        imgs_3 = [Image.open(p).convert('RGB') for p in pngs_3]
        imgs_3[0].save('outputs/handout/salary_handout_negotiation.pdf', save_all=True, append_images=imgs_3[1:])
        print(f'✅ PDF 3: salary_handout_negotiation.pdf ({len(pngs_3)} pages)')
    
    pdf_count = sum([bool(pngs_1), bool(pngs_2), bool(pngs_3)])
    print(f'\n📄 Generated {pdf_count} PDF dashboards in outputs/handout/')
    
except ImportError:
    print('⚠️  PIL/Pillow not installed - PDFs not created')
except Exception as e:
    print(f'⚠️  PDF generation error: {e}')
```
